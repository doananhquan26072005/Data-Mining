```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn import preprocessing
from sklearn.decomposition import PCA

feature_names = [f'f{i}' for i in range(1, 167)]
column_names = ['molecule_name', 'conformation_name'] + feature_names + ['class']

df = pd.read_csv('D:\\data_mining\\tuan1\\clean2.data', header= None, names=column_names)
print(df)
```

```python
#2 - Kiểm tra các giá trị missing
df.replace('?', np.nan, inplace=True)
print(df.isnull().values.any())
```

```python
#Kiểm tra giá trị giống nhau
print(df.duplicated().sum())
```

```python
X = df.drop(['molecule_name', 'conformation_name', 'class'], axis=1)
#- chuẩn hóa dữ liệu bằng phương pháp z-score(Standard)
scaler = preprocessing.StandardScaler()
array = scaler.fit_transform(X)

df_standard = pd.DataFrame(array, columns=X.columns)

print(df_standard.head())
```

```python
#Áp dụng PCA
pca = PCA() 
pca.fit(df_standard)
```

```python
eigenvalues = pca.explained_variance_
eigenvectors = pca.components_

for i, val in enumerate(eigenvalues):
    print(f"Thành phần chính {i+1}:")
    print(f"Eigenvalue (Trị riêng): {val:.4f}")
    print(f"Eigenvector (Vector riêng): {eigenvectors[i]}")
    print("-" * 30)
```
