```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('D:\\data_mining\\tuan2\\kddcup.data.corrected', header=None)
print(df)
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
print("Categorucal attributes:", categorical_columns)
categorical_data = df[categorical_columns]
print("Categorical Data:\n", categorical_data)
```

```python
#- Kiểm tra giá trị giống nhau
print("So luong dong giong nhau:", categorical_data.duplicated().sum())

##loại bỏ các dòng giống nhau
df1 = categorical_data.drop_duplicates()
print("Du lieu sau khi xoa nhung dong giong nhau: \n", df1)
```

```python
array = df1.values
array = array[: 100]
array = np.array(array)
```

```python
def Overlap(X, Y):
    sim = 0
    for i in range(len(X)):
        if X[i] == Y[i]:
            sim = sim + 1/len(X)
    return sim
```

```python
def tan_suat_xuat_hien_nguoc(X, Y):
    sim = 0
    for i in range(len(X)):
        if X[i] == Y[i]:
            sim = sim + 1/len(X)
        if X[i] != Y[i]:
            count1 = np.sum(array[:, i] == X[i])
            count2 = np.sum(array[:, i] == Y[i])
            sim = sim + 1/len(X) * 1/(1 + np.log(count1) * np.log(count2))
    return sim
```

```python
Overlap_array = np.zeros((100, 100))
tsxhn_array = np.zeros((100, 100))

for i in range(0 , 100):
    for j in range(0, 100):
        Overlap_array[i][j] = Overlap(array[i], array[j])
        tsxhn_array[i][j] = tan_suat_xuat_hien_nguoc(array[i], array[j])

print(Overlap_array, "\n")
print(tsxhn_array)
```

```python
neighbor_Overlap = np.argsort(-Overlap_array, axis=1)
neighbor_tsxhn = np.argsort(-tsxhn_array, axis=1)

print(neighbor_Overlap, "\n")
print(neighbor_tsxhn)
```
