```python
#array
array = df.values
print(array)

point1 = array[0,:]
point2 = array[1,:]
point3 = array[2,:]

#p = 1
dist01_2 = np.linalg.norm(point1 - point2, 1)
dist01_3 = np.linalg.norm(point1 - point3, 1)

#p = 2
dist1_2 = np.linalg.norm(point1 - point2, 2)
dist1_3 = np.linalg.norm(point1 - point3, 2)

#p = inf
dist11_2 = np.linalg.norm(point1 - point2, np.inf)
dist11_3 = np.linalg.norm(point1 - point3, np.inf)

print(dist01_2)
print(dist01_3)

print(dist1_2)
print(dist1_3)

print(dist11_2)
print(dist11_3)
```

```python
def Distanace(X, Y, p):
    if p == np.inf:
        return max(np.abs(X - Y))
    return np.sum(np.abs(X - Y) ** p) ** (1 / p)
```

```python
Dist_1 = np.zeros((50, 50))
Dist_2 = np.zeros((50, 50))
Dist_inf = np.zeros((50, 50))

for i in range(0 , 50):
    for j in range(0, 50):
        Dist_1[i][j] = Distanace(array[i], array[j], 1)
        Dist_2[i][j] = Distanace(array[i], array[j], 2)
        Dist_inf[i][j] = Distanace(array[i], array[j], np.inf)

print(Dist_1, "\n")
print(Dist_2, "\n")
print(Dist_inf, "\n")
```
