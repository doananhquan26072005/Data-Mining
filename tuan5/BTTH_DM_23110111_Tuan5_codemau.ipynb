```python
import numpy as np
import pandas as pd
from pyECLAT import ECLAT

dataframe = pd.read_csv(r"D:\data_mining\tuan4\data.csv", header=None)
print(dataframe)
```

```python
eclat_instance = ECLAT(data=dataframe, verbose=True)
print(eclat_instance.df_bin)
```

```python
# count items in each row
items_per_transaction = eclat_instance.df_bin.astype(int).sum(axis=1)
# the item should apear at least at 5% of transactions
min_support = 0.5
# start from tracsactions containing at least 2 items
min_combination = 2
# up to maximum items per transaction
max_combination = max(items_per_transaction)
rule_indeces, rule_supports = eclat_instance.fit(min_support=min_support,
                                                 min_combination=min_combination,
                                                 max_combination=max_combination,
                                                 separator=' & ',
                                                 verbose=True)
results = pd.DataFrame(rule_supports.items(), columns=['Item', 'Support'])
results.sort_values(by=['Support'], ascending=False)
print(results)
```
